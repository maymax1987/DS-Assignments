```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv("Online retail1.csv")         
```

```python
df
```

# Association Rule Mining:

```python
df.drop_duplicates(inplace=True)
```

```python
def txt_split(txt):
    return txt.split(',')
```

```python
data = [i.split(',') for i in df['products']]
data
```

```python
te = TransactionEncoder()
encoded_df = te.fit_transform(data)
```

```python
encoded_df
```

```python
te.columns_
```

```python
data = pd.DataFrame(encoded_df, columns=te.columns_)
data
```

```python
data.replace([True, False],[1,0], inplace=True)
data
```

# Analysis and Interpretation:

```python
scores = apriori(data,min_support=0.05, use_colnames=True)
scores
```

```python
rules = association_rules(scores, metric="lift", min_threshold=1)
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

```python
rules.sort_values(by = 'lift', ascending=False)
```
